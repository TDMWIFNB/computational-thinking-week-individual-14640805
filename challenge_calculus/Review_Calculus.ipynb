# Silver medal: Find the derivative of a function using the product rule and the rule of composite function

a) Student B carefully examined his data and checked the literature. After that, he use this function to fit his data:

## $y=144000\cdot e^{-x/8}\left( -x - 8\right)+1152000$

where x is the past time in hour and y is the total viewing time (in the unit of minus). 

Can you find the deriviative of this function?

First, we need this formula to actually be a function by replacing the y by f(x)
For this, it makes the most sense to pull the (-x-8) to the front, so this one can be multiplied with the number in front. This creates the following function: f(x) = (-144000x - 1152000)e^(-x/8) + 1152000.
Following this, the product rule can be used to get the derivative of the function.
Via it's specifications, the following solution follow: f'(x) = -144000 * e^(-x/8) + (-144000x - 1152000) * (-1/8) * e^(-x/8) = -144000 * e^(-x/8) + 18000x * e^(-x/8) + 144000 * e^(-x/8) = 18000x * e^(-x/8).


Thereby, f'(x) = 18000x * e^(-x/8).



b) Following their previous discussion, assuming the correctness of the function found by Student B, they should be able to find the function that the number of clicks per hour will obey. Can you help them find this function?

c) Use the best model to calculate: at which hour does clicks per hour reach its maximum value? (rounded to two decimal places)

```python
# Start to calculate the time (hour) for the highest value

import scipy

# Source: https://stackoverflow.com/questions/10146924/finding-the-maximum-of-a-function
def find_max_clicks(model):
    maxs = scipy.optimize.fmin(lambda x: -model(x), 0)
    max = maxs[0]
    pass
    return max

# Print the maximum clicks per hour
print("The maximum clicks per hour occur at {:.2f}.".format(find_max_clicks(BestModel)))

# I checked the end-value by finding the derivative of the BestModel and afterwards calculating the value where y = 0 here. This was x=8 as well.
```

d) This YouTuber happened to get a double-coupon. This special double-coupon can doubles his profits from all viewings for 5 hours and he can choose the exact time to start. Based on the model we found, can you suggest which hour he should start using this coupon? The time is counted as 0 when his video is published. (rounded to two decimal places)

```python
from scipy.integrate import quad
# Start to calculate the best time (hour) for the coupon
# You can use places outside of this function, including defining new parameters and new functions.
def find_best_coupon_time(model):
    # reusing the function I used in exercise c to get the maximum of the function
    maxs = scipy.optimize.fmin(lambda x: -model(x), 0)
    max = maxs[0]

    max_area = 0
    best_start = None

    # creating a for loop to calculate the biggest area underneath the graph surrounding the maximum (because the maximum certainly is in there when looking at the model)
    # I used 0.001 for my step size, due to the possibility a rounding error would have appeared at 0.01
    for x in np.arange(max - 5, (max + 5) + 0.001, 0.001):
        # defining variables for limits
        lower_limit = x
        upper_limit = x + 5
        
        # making sure the limit does not exceed over the boundary after the maximum
        if upper_limit > max + 5:
            upper_limit = max + 5
        
        # calculating the area underneath the graph for a certain combination of upper and lower limits
        area, error = quad(model, lower_limit, upper_limit)
        
        # change the maximum area if it is bigger then the previously calculated biggest area 
        if area > max_area:
            max_area = area
            best_start = lower_limit
    
    pass
    # returning the lower limit of the values used to calculate the biggest area in the interval
    return best_start

# Finding best 5-hour period to activate the double coupon
print("The best time to start the 5-hour double coupon is at hour {:.2f}".format(find_best_coupon_time(BestModel)))
```
